```python
# For the date early in the dataset, measure the arbitrage spread between the two exchanges
# by subtracting the lower-priced exchange from the higher-priced one

# Use a conditional statement to generate the summary statistics for each arbitrage_spread DataFrame
# YOUR CODE HERE

arbitrage_spread_early = coinbase['Close'].loc['2018-01-07'] - bitstamp['Close'].loc['2018-01-07']

arbitrage_spread_early = arbitrage_spread_early[arbitrage_spread_early > 0]

arbitrage_spread_early.describe()
```

```python
arbitrage_spread_middle = coinbase['Close'].loc['2018-02-01'] - bitstamp['Close'].loc['2018-02-01']

arbitrage_spread_middle = arbitrage_spread_middle[arbitrage_spread_middle > 0]

arbitrage_spread_middle.describe()
```

```python
arbitrage_spread_late = coinbase['Close'].loc['2018-03-25'] - bitstamp['Close'].loc['2018-03-25']

arbitrage_spread_late = arbitrage_spread_late[arbitrage_spread_late > 0]

arbitrage_spread_late.describe()
```

#### 2. For each of the three dates, calculate the spread returns. To do so, divide the instances that have a positive arbitrage spread (that is, a spread greater than zero) by the price of Bitcoin from the exchange you’re buying on (that is, the lower-priced exchange).
```python
# For the date early in the dataset, calculate the spread returns by dividing the instances when the arbitrage spread is positive (> 0) 
# by the price of Bitcoin from the exchange you are buying on (the lower-priced exchange).
print(early)
spread_return_early = arbitrage_spread_early/bitstamp['Close'].loc[early]

# Review the spread return DataFrame
# YOUR CODE HERE
spread_return_early.to_frame()
```
